```python
from tprof import load_data
from tprof import tprof

#Making a temperature-pressure profile, from a GCM output
#need fort.26, fort.50, in the same folder 

#loading in the data
path='/Users/hbeltz/Documents/runs/' #where the fort files are stored
runname='test03' #goes into another folder, leave as '' if no need to go into another folder
oom=7 #orders of mangitude covered in the model
surfp=100 #surface pressure, in bars
verbose=False #would you like a bunch of additional information? 
fort_26= 'fort.2615'
nlay=45     #number of levels

runname,lon_arr,lat_arr,oom,surfp,p_BAR,data_26=load_data(path,runname,oom,surfp,verbose,fort_26)




```

```python
#making the plot
savefig=False #would you like to save the figure?
savepath='/Users/hbeltz/Documents/runs/' #choosing where to save the figure
runname='test03'
savename=savepath+runname+'/'+runname+'tprofnewfinal'+'.png' #choosing the name


%matplotlib inline
tprof(lon_arr,lat_arr,p_BAR,data_26,oom,surfp,nlay,runname,path,
         savefig,savename)
```
