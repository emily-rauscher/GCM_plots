```python
import numpy as np

import sys
#print (sys.argv)
sys.path.insert(0,'Main_Functions/') 

from FortranFileFinisher import *

#######################################
#     NECESSARY USER INPUT BELOW!     #
#######################################

# put the main path to your runs here
path='/Users/ermay/Documents/GradSchool/Research_Surface/GCM_plots/Files/'
runname='EARTH_T42'

# don't forget to update the inputs below!
# this is important, fort26 saves the lon,lat, and p_BAR arrays for later use! 
# if they're wrong now, you'll have to remake these files by rerunning this cell
oom=0
surfp=1.00
radea=6.37*10**6
base_flux=0.4
nlev=30
nday=1800

# do you want it to print the full Longitude, Latitude, and Pressure arrays? (probs not...)
ver=False

# do you want to save the converted fortan file to a pandas dataframe? (probs yes...)
savet=False  #you don't want to save for last orbit averages!

# the last input lets you load in a specific one of the fort.* files from the last orbit if you'd like
# FYI: fort65, 66 are from new versions of the GCM not yet pushed and merged to bitbucket. 
# you can comment them out if you don't have them yet
# -----------------------------------------------
# fort.26: Winds, Temperatures @all lats,lons,levs
# fort.50: forthcoming
# fort.64,65: Emitted LW,SW,and Total TOA Fluxes, respectively
# (fort.66 currently doesn't support final orbit averaging)
# -----------------------------------------------

prs_load = pickle.load(open(path + runname + '/pres_lon_lat.txt', 'rb'))[0]
lon_load = pickle.load(open(path + runname + '/pres_lon_lat.txt', 'rb'))[1]
lat_load = pickle.load(open(path + runname + '/pres_lon_lat.txt', 'rb'))[2]

nlev,nlon,nlat=len(prs_load),len(lon_load),len(lat_load)

full_26 = np.empty([90,nlev,nlon,nlat,6])*np.nan
full_64 = np.empty([90,nlev,nlon,nlat])*np.nan
full_65 = np.empty([90,nlev,nlon,nlat])*np.nan

filen=0
while filen<90:
    file_26='fort.'+str(int(2600+filen))
    file_64='fort.'+str(int(6400+filen))
    file_65='fort.'+str(int(6500+filen))
    
    print ' '
    print '--------------------------'
    print '------> OUTPUT #:', filen

    runname, oom, surfp, lon_arr, lat_arr, p_BAR, full_26[filen,:,:,:,:] = fort26(path,runname,oom,surfp,ver,savet,file_26)
    _______,             _______, _______, full_64[filen,:,:,:], tt_lw = fort64(path,runname,radea,ver,savet,file_64)
    _______,             _______, _______, full_65[filen,:,:,:], tt_sw = fort65(path,runname,radea,ver,savet,file_64)

    filen+=1
    
pickle.dump(full_26[:,:,:,:,3], open(path+runname+'/UWD_lo.txt','wb'))
pickle.dump(full_26[:,:,:,:,4], open(path+runname+'/VWD_lo.txt','wb'))
pickle.dump(full_26[:,:,:,:,5], open(path+runname+'/TMP_lo.txt','wb'))
pickle.dump(full_64, open(path+runname+'/LWF_lo.txt','wb'))
pickle.dump(full_65, open(path+runname+'/LWF_lo.txt','wb'))

pickle.dump(np.nanmean(np.nanmean(full_26[:,:,:,:,3],axis=2),axis=0), open(path+runname+'/UWD_avrg.txt','wb'))
pickle.dump(np.nanmean(np.nanmean(full_26[:,:,:,:,4],axis=2),axis=0), open(path+runname+'/VWD_avrg.txt','wb'))
pickle.dump(np.nanmean(np.nanmean(full_26[:,:,:,:,5],axis=2),axis=0), open(path+runname+'/TMP_avrg.txt','wb'))
pickle.dump(np.nanmean(np.nanmean(full_64,axis=2),axis=0), open(path+runname+'/LWF_avrg.txt','wb'))
pickle.dump(np.nanmean(np.nanmean(full_65,axis=2),axis=0), open(path+runname+'/LWF_avrg.txt','wb'))
```
